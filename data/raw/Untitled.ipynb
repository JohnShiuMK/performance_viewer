```python
import pandas as pd
import pytz
```

```python
df = pd.read_csv("BTCUSDT-1m-2024-04-21.csv")
df['datetime_utc'] = pd.to_datetime(df['close_time'], unit='ms', utc=True)
df
```

```python
df['cummax'] = df['close'].cummax()
df['drawdown'] = df['close'] / df['cummax'] - 1
df['return'] = (df['close'] / df['close'].shift() - 1).round(4)
df.head()
```

```python
annual_return = (df['return'] * (60 * 24 * 365)).mean()
annual_risk = (df['return'] * (60 * 24 * 365)**0.5).std()
max_dd = -df['drawdown'].min()
max_dd
calmar = annual_return / max_dd
calmar
```

```python
pd.DataFrame({
    'Strategy': 'test',
    'Annualized Return': annual_return,
    'Annualized Risk': annual_risk,
    'Maximum Drawdown': max_dd,
    'Sharpe Ratio': annual_return / annual_risk,
    'Calmar Ratio': annual_return / max_dd
}, index=[0])
```
